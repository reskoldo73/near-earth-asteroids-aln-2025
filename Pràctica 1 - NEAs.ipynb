# Pràctica 1: NEAs

## **Introducció**

El sistema solar és un sistema planetari format pel Sol i una gran varietat d'objectes celestes que orbiten al seu voltant. Aquests objectes inclouen els vuit planetes principals (Mercuri, Venus, Terra, Mart, Júpiter, Saturn, Urà i Neptú), els seus satèl·lits, planetes nans, cometes i asteroides. Els asteroides són cossos rocosos més petits que els planetes, principalment situats al cinturó d'asteroides entre Mart i Júpiter. Actualment coneixem aproximàdmanet un milió i mig d'asteroides al sistema solar, amb diàmetres que van des de uns pocs metres fins a centenars de quilòmetres.

Dins d'aquesta gran població d'asteroides, hi ha un subconjunt conegut com a **Asteroides Propers a la Terra** (NEAs, per les seves sigles en anglès: *Near-Earth Asteroids*). Aquests, són asteroides que assoleixen distàncies inferiors a 1.3 unitats astronòmiques (AU) i que, per tant, passen aprop de l'òrbita de la Terra. Més concretament, la **Unitat Astronòmica** (UA, en anglès *Astronomical Unit*, AU) és una unitat de longitud utilitzada principalment en astronomia per mesurar distàncies dins del sistema solar. Es defineix com la distància mitjana entre la Terra i el Sol i el seu valor és $1 \, \text{UA} = 149\,597\,870\,700\, \text{m}$.

En l'actualitat coneixem gairebé 40 000 NEAs[$^{[1]}$](https://cneos.jpl.nasa.gov/stats/totals.html) i podem veure la majoria de la trajectòria d'aquests al següent enllaç[$^{[2]}$](https://science.nasa.gov/solar-system/asteroids/).

### **Importància de l'estudi dels NEAs**

L'estudi dels NEAs és crucial per diverses raons:

* **Protecció Planetària**: Comprendre les òrbites i les característiques físiques dels NEAs ens permet avaluar el risc d'impacte amb la Terra i desenvolupar estratègies per mitigar aquest risc.

* **Ciència Bàsica**: Els NEAs són restes de la formació del sistema solar, i estudiar-los ens proporciona informació valuosa sobre les condicions i processos que van donar lloc als planetes.

* **Exploració Espacial**: Alguns NEAs són objectius potencials per a missions espacials, tant per a l'exploració científica com per a la possible extracció de recursos minerals.

### **Dinàmica dels NEAs**

La dinàmica dels NEAs es pot aproximar inicialment utilitzant el **problema de Kepler**, que descriu el moviment d'un objecte sota la influència gravitatòria d'un altre cos massiu, com el Sol. Aquesta aproximació és vàlida a curt termini, ja que el Sol representa més del 99.86% de tota la massa del sistema solar. Així, considerant un problema de on el Sol exerceix de massa central podem predir, a curt termini, les òrbites dels NEAs amb una precisió raonable.

Si considerem un sistema de refererència Heliocèntric (és a dir, el Sol està situat a l'origen de coordenades) tenim que l'equació del moviment del NEA es pot aproximar per

$$
\frac{d^2 \boldsymbol{r}}{dt^2} = -\frac{\mu}{r^3}\boldsymbol{r},
$$

on $\boldsymbol{r}$, denota la posició del NEA, $r=|\boldsymbol{r}|$ és la distància del NEA al Sol i $\mu = GM$ és el **paràmetre gravitacional estàndard** del Sol. Aquesta equació descriu una òrbita kepleriana, que pot ser circualr, el·líptica, parabòlica o hiperbòlica, depenent de l'energia total del sistema.

&nbsp;

No obstant això, a llarg termini, la dinàmica dels NEAs és molt més complexa. Això es deu a la influència gravitatòria d'altres cossos del sistema solar, com els planetes (especialment Júpiter), que poden pertorbar les seves òrbites de manera significativa i altres efectes com poden ser la pressió de radiació Solar o l'efecte Yarkovsky. Aquestes pertorbacions poden causar canvis substancials en les òrbites dels NEAs, fent que la seva trajectòria a llarg termini sigui difícil de modelar amb precisió.

Per aquest motiu, tenim que les trajectòries dels NEAs i, en general dels asteroides, es poden aproximar per una cònica, tot i que a la pràctica no ho són exactament.

### **Elements orbitals:**

Els **elements orbitals** són un conjunt de paràmetres que descriuen de manera única la forma, la mida, l'orientació i la posició d'una òrbita en l'espai. Aquests elements són essencials per a la mecànica celeste, ja que permeten definir i predir el moviment d'un objecte al voltant d'un altre sota la influència de la gravetat. Per a una òrbita kepleriana (el·líptica, parabòlica o hiperbòlica), es necessiten sis elements orbitals per descriure completament l'òrbita:

* **Semieix major ($a$)**:
   - Defineix la mida de l'òrbita. En una òrbita el·líptica, és la meitat de la longitud de l'eix major.
   - Per a òrbites circulars, el semieix major és igual al radi de l'òrbita.

* **Excentricitat ($e$)**:
   - Descriu la forma de l'òrbita. Pot prendre valors entre 0 i 1 per a òrbites el·líptiques:
     - $e = 0$: òrbita circular.
     - $0 < e < 1$: òrbita el·líptica.
     - $e = 1$: òrbita parabòlica (cas límit).
     - $e > 1$: òrbita hiperbòlica.

* **Inclinació ($i$)**:
   - Defineix l'angle entre el pla de l'òrbita i un pla de referència (generalment el pla de l'eclíptica per a objectes del sistema solar).
   - Pot variar entre $0^\circ$ i $180^\circ$.

* **Longitud del node ascendent ($\Omega$)**:
   - Defineix l'orientació de l'òrbita respecte al punt de referència (generalment el punt d'Àries o punt vernal[$^{[3]}$](https://ca.wikipedia.org/wiki/Punt_d%27%C3%80ries)).
   - És l'angle entre el punt vernal i el node ascendent (el punt on l'òrbita creua el pla de referència de sud a nord).

* **Argument del periheli ($\omega$)**:
   - Descriu l'orientació de l'òrbita dins del seu pla.
   - És l'angle entre el node ascendent i el periheli (el punt de l'òrbita més proper al Sol).

* **Anomalia mitjana ($M$)**:
   - Defineix la posició de l'objecte en l'òrbita en un moment donat.
   - Es relaciona amb el temps mitjançant l'equació de Kepler.

Utilitzant Gravity Simulator[$^{[4]}$](https://gravitysimulator.org/solar-system/the-solar-system) es pot obtenir una idea més intuitiva d'aquests.

### **Exercici 2: Període orbital**

Un cop hem resolt l'Exercici 1, estimarem el període orbital de l'asteroide utilitzant dos mètodes diferents.

#### **Primer mètode: Mínims relatius de distància al Sol**

El primer mètode consisteix a calcular els instants en què l'asteroide passa pel seu periheli, és a dir, els moments en què la distància al Sol és mínima. Aquests mínims es detecten quan la distància $r_i$ en un instant $t_i$ és menor que les distàncies als instants adjacents ($r_{i-1}$ i $r_{i+1}$).

> ⚠️ **Atenció!**
>
> No es consideren els extrems de l'interval de temps, és a dir, no es tenen en compte $t_0$ ni $t_{\text{final}}$.

Un cop identificats tots els instants $t_{k_1}, t_{k_2}, \dots, t_{k_n}$ en què es produeixen aquests mínims, el període orbital $T_*$ es calcula com la mitjana del temps entre aquests instants:
$$
T_* = \frac{t_{k_n} - t_{k_1}}{n - 1}
$$

#### **Segon mètode: Elements orbitals**

El segon mètode utilitza els elements orbitals obtinguts en el primer apartat. A partir d'aquests, el període orbital $T$ es calcula mitjançant la següent fórmula:
$$
T = 2\pi \sqrt{\frac{a^3}{GM}}
$$
on:
- $a$ és el semieix major de l'òrbita,
- $GM$ és una constant que val aproximadament $2.959122082856\cdot10^{-4}\, \text{AU}^3/\text{days}^2$.

### Exercici 3: Distància mínima a la Terra

En aquest exercici, calcularem la distància mínima entre un asteroide i la Terra, així com l'instant exacte en què es produeix aquesta distància mínima. Per fer-ho, seguirem un procés basat en l'anàlisi de les dades de posició de la Terra i l'asteroide.

&nbsp;

Primer, obteniu les dades de la posició de la Terra en els mateixos instants de temps que les de l'asteroide utilitzant la funció `getEphem("399")`. Aquesta funció us proporcionarà les coordenades de la Terra en cada instant $t_i$. Si $\boldsymbol{q}_i$ representa la posició de la Terra i $\boldsymbol{r}_i$ la posició de l'asteroide en l'instant $t_i$, podeu calcular la distància entre ells mitjançant la fórmula $d_i = \|\boldsymbol{q}_i - \boldsymbol{r}_i\|_2$.

&nbsp;

Un cop calculades totes les distàncies $d_i$, identifiqueu l'índex $i$ on aquesta distància és mínima. A continuació, preneu els valors de distància propers a aquest mínim ($d_{i-3}, d_{i-2}, \dots, d_{i+3}$) i construiu una funció quadràtica que aproximi la distància al voltant d'aquest punt. La funció té la forma

$$D(t - t_i) = D_0 + D_1 (t - t_i) + D_2 (t - t_i)^2,$$

on $D_0$, $D_1$ i $D_2$ són constants que haureu de determinar mitjançant un ajust per mínims quadrats.

&nbsp;

Per realitzar aquest ajust, implementeu una funció que faci la factorització QR emprant el mètode de Gram-Schmidt modificat. Aquest mètode us permetrà resoldre el sistema d'equacions de manera eficient per trobar les constants de la funció quadràtica.

&nbsp;

Finalment, un cop determinada la funció $D(t - t_i)$, calculeu la distància mínima a la Terra i l'instant de temps $t$ en què es produeix aquesta distància mínima.

### Funció **`getEphem`**:

Ens permet obtenir les dades efemèrides mitjançant l'API de JPL Horizons i les guarda en un fitxer csv:

````python
import pandas as pd
import requests

def getEphem(idObjecte, startDate="2025-MAY-01", endDate = "2051-JAN-01", step = 12):
    """
    Obté les dades efemèrides d'un objecte astronòmic mitjançant l'API de JPL Horizons
    i les guarda en un fitxer CSV.

    Paràmetres:
    -----------
    idObjecte : str
        Identificador de l'objecte astronòmic (per exemple, "199" per a Venus o "301" per a la Lluna).
    startDate : str, opcional
        Data d'inici per a les efemèrides (per defecte: "2025-MAY-01").
    endDate : str, opcional
        Data de fi per a les efemèrides (per defecte: "2051-JAN-01").
    step : int, opcional
        Interval de temps entre les dades en hores (per defecte: 12).

    Retorna:
    --------
    None
        La funció no retorna cap valor, però guarda les dades en un fitxer CSV amb el nom de l'objecte.

    Exemple:
    --------
    >>> getEphem("199", startDate="2025-JUL-01", endDate="2025-JUL-10", step=24)
    Això generarà un fitxer anomenat "199.csv" amb les dades efemèrides de Venus des de l'1 de juliol de 2025
    fins al 10 de juliol de 2025, amb un interval de 24 hores entre les dades.

    C. Óscar Rodríguez
    """

    # Generem la URL per fer la consulta API
    url = (
        "https://ssd.jpl.nasa.gov/api/horizons.api?format=text"
        "&MAKE_EPHEM='YES'"
        f"&COMMAND='{idObjecte}'"     # Objecte
        "&EPHEM_TYPE='VECTORS'"
        "&CENTER='500@10'"
        f"&START_TIME='{startDate}'"  # Data inicial
        f"&STOP_TIME='{endDate}'"     # Data final
        f"&STEP_SIZE='{step} HOURS'"  # Distància entre les dades (en hores)
        "&VEC_TABLE='2'"
        "&REF_SYSTEM='ICRF'"
        "&REF_PLANE='ECLIPTIC'"
        "&VEC_CORR='NONE'"
        "&CAL_TYPE='M'"
        "&OUT_UNITS='KM-S'"
        "&VEC_LABELS='YES'"
        "&VEC_DELTA_T='NO'"
        "&CSV_FORMAT='YES'"
        "&OBJ_DATA='YES'"
    )

    # Obtenim les dades:
    response = requests.get(url)
    response.raise_for_status() # Excepció si falla

    # Decodifiquem el contingut a una string:
    text = response.content.decode('utf-8')

    # Trobem l'índex on comencen i acaben les dades:
    start = text.find('$$SOE') + 5
    end = text.find('$$EOE')

    # Formatem les dades:
    data = "MJD, UTC,x, y, z, vx, vy, vz," + text[start:end]
    data = data.replace(',\n', '\n')

    # Guardem en un CSV les dades:
    with open(f"{idObjecte.replace('DES=','')}.csv", 'w') as fitxer:
       fitxer.write(data)
````

### Funció **`car2kep`**:

Ens permet obtenir les coordenades keplerianes a partir de les coordenades cartesianes (posicions i velocitats) en un sistema de referència eclíptic i heliocèntric.

Pe vostre cas únicament heu de fer-la servir sense utilitzar el paràmetre opcional K.

````python
def car2kep(x, K = 132712440041.93938):
  """
  car2kep(x, K = 132712440041.93938) transforma les coordenades cartesianes d'una
     òrbita el·líptica heliocèntrica als seus elements orbitals keplerians.

  INPUT:
    - x: array unidimensional que conté la posició (x,y,z) i velocitat (v_x,v_y,v_z)
         de l'objecte.
    - K: paràmetre opcional que conté el quadrat de la constant de Gauss en les
         unitats adients.
  OUTPUT:
    - array unidimensional que conté, per ordre, els valors de:
        - a (semieix major [AU])
        - e (excentricitat)
        - i (inclinació [graus])
        - Omega (longitud del node ascendent [graus])
        - omega (argument del periàpside [graus])
        - M (anomalia mitjana [graus])

  C. Óscar Rodríguez, 2024
  """

  # Posició, velocitat i moment angular:
  r = x[0:3]
  v = x[3:6]
  c = np.cross(r,v)

  # Norma euclidiana de la posició, la velocitat i el moment angular.:
  rnorm = np.linalg.norm(r)
  vnorm = np.linalg.norm(v)
  cnorm = np.linalg.norm(c)

  # Vector excentricitat i excentricitat
  ecc = ((vnorm*vnorm-K/rnorm)*r - np.dot(r,v)*v)/K;
  e = np.linalg.norm(ecc)

  # Vector n i la seva norma:
  n = np.array([-c[1], c[0], 0])
  nnorm = np.linalg.norm(n)

  # Anomalia verditable:
  ang = np.max([np.min([np.dot(ecc,r)/(e*rnorm),1]),-1]) # ang\in[-1,1]
  nu = np.arccos(ang)
  if np.dot(r,v) < 0:
    nu = 2*np.pi-nu

  # Inclinació:
  i = np.arccos(c[2]/cnorm)

  # Anomalia excèntrica:
  E = 2*np.arctan(np.tan(nu/2)/np.sqrt((1+e)/(1-e)))

  # Anomalia mitjana:
  M = E-e*np.sin(E)

  # Longitud del node ascendent:
  Omega = np.arccos(n[0]/nnorm)
  if n[1] < 0:
    Omega = 2*np.pi - Omega

  # Argument del periàpside:
  omega = np.arccos(np.dot(n,ecc)/(nnorm*e))
  if ecc[2] < 0:
    omega = 2*np.pi - omega

  # Semi-eix major:
  a = 1/(2/rnorm - vnorm*vnorm/K)

  return np.array([a/149597870.7,     # semi-eix major (AU)
                   e,                 # excentricitat
                   np.rad2deg(i),     # inclinació (graus)
                   np.rad2deg(Omega), # longitud del node ascendent (graus)
                   np.rad2deg(omega), # argument del periàpside (graus)
                   np.rad2deg(M)      # anomalia mitjana (graus)
                  ])
````

### Codi per dibuixar un ajust per mínis quadrats

Suposem que tenim un conjunt de dades $(x_1,y_1)$, $(x_2,y_2)$, $\ldots$, $(x_m,y_m)$ i que hem construit el polinomi $p(x) = p_0 + p_1 x +\ldots + p_n x^n$ que millor aproxima (per mínims quadrats) aquestes dades.

Suposem a més, que tenim aquesta informació emmagatzemada en tres vectors:

$$\begin{aligned}\boldsymbol{x} &= (x_1,x_2,\ldots,x_m),\\\boldsymbol{y} &= (y_1,y_2,\ldots,y_m),\\\boldsymbol{p} &= (p_1,p_2,\ldots,p_n).\end{aligned}$$

La funció `plotMinQuad` ens permet dibuixar les dades $(x_1,y_1)$, $(x_2,y_2)$, $\ldots$, $(x_m,y_m)$ així com el polinomi $p(x)$ en els punts $x\in[\min(\boldsymbol{x}),\max(\boldsymbol{x})]$.

````python
import numpy as np
import matplotlib.pyplot as plt

def plotMinQuad(x,y,p,xaxis="x",yaxis="y",title="Ajust mínims quadrats"):
    """
    Dibuixa un gràfic que mostra un conjunt de dades i un polinomi ajustat mitjançant el mètode dels mínims quadrats.

    Paràmetres:
    -----------
    x : array-like
        Array o llista que conté les coordenades x dels punts de dades.
    y : array-like
        Array o llista que conté les coordenades y dels punts de dades.
    p : array-like
        Array o llista que conté els coeficients del polinomi ajustat, ordenats de menor a major grau.
    xaxis : str, opcional
        Etiqueta per a l'eix x del gràfic (per defecte: "x").
    yaxis : str, opcional
        Etiqueta per a l'eix y del gràfic (per defecte: "y").
    title : str, opcional
        Títol del gràfic (per defecte: "Ajust mínims quadrats").

    Exemple d'ús:
    ------------
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
    p = [0.98, 1.02]  # Coeficients del polinomi ajustat
    plotMinQuad(x, y, p, xaxis="Temps (s)", yaxis="Distància (m)", title="Ajust lineal")

    C. Óscar Rodríguez
    """

    # Número de punts on avaluem el polinomi:
    N = 500

    # Creem un vector amb punts equipespaiat entre el valor més petit de x i el més gran:
    xs = np.linspace(np.min(x),np.max(x),N)

    # Avaluem el polinomi aquests punts utilitzant la regla de Horner:
    pxs = p[-1]
    for i in range(2,len(p)+1):
        pxs = pxs*xs + p[-i]

    # Dibuixem els punts i el polinomi:
    plt.plot(xs,pxs,label=f"$p(x)$",color='blue')
    plt.scatter(x,y,label="Dades",s=5,color='black')
    plt.xlabel(xaxis)
    plt.ylabel(yaxis)
    plt.title(title)
    plt.legend()
    plt.show()



````

Veiem un exemple:

````python
# Exemple de dades:
x = np.linspace(-1,1, 6)
y = np.array([1.989,0.7,0.1,0.1,0.7,1.989])
p = np.array([0,0,2])

# Dibuixem:
plotMinQuad(x,y,p)
````
